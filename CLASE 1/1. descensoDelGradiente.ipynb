```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, deriv=False):
    if (deriv==True):
        return (2/5)*np.sin(x**3)*np.cos(x**3)*(3*x**2)
    return (1/5)*(np.sin(x**3)**2)
```

```python
# learning rate
lr = .05

X = []
y = []
```

```python
x = 1
while (f(x) > .001):
    x -= lr*f(x, deriv=True)
    X.append(x)
    y.append(f(x))
```

```python
print('f(x) = ', f(x))
print('x = ', x)
print("f'(x) = ", f(x, deriv = True)) 
```

```python
xx = np.linspace(-2, 2, 100)
plt.plot(xx, f(xx))
plt.plot(X, y, '.', c = 'red')
plt.plot(X[len(X) - 1], y[len(y) - 1], 'o', c = 'gold')
plt.show()
```
