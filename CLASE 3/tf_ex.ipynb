```python
import numpy as np
import NNframework as NN
```

```python
x = np.array([[-2, -3, -1, 1, 0, 2, 4, 5, 6]]).T
y = np.array([[-5, -7, -3, 1, -1, 3, 7, 9, 11]]).T
```

```python
nn = NN.NeuralNetwork()
```

```python
nn.createNetwork([1, 1], ['linear'])
```

```python
nn.train(x, y, 0.005, decay=0.001, epochs=1500)
```

```python
nn.print()
```
