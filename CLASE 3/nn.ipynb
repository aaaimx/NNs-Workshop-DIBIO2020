```python
import numpy as np
import NNframework as NF
```

```python
x = [
    [-1, 1],
    [-1, -1],
    [1, 1],
    [1, -1]
]

y = [
    [1, 1],
    [1, -1],
    [1, 1],
    [-1, 1]
]
```

```python
nn = NF.NeuralNetwork()
```

```python
nn.createNetwork([2, 3, 2], ['tanh', 'tanh'])
```

```python
nn.train(x, y, 0.005, decay=0.001, epochs=10000)
```

```python
nn.predict([-1, -1])
```
